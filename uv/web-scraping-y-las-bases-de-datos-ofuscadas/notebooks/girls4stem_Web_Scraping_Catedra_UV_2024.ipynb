```python
import google_colab_selenium
from selenium.webdriver.common.by import By
from selenium import webdriver
```

```python
options = webdriver.ChromeOptions()
options.add_argument("--headless")
options.add_argument("--no-sandbox")
options.add_argument("--disable-dev-shm-usage")
# options.add_argument(f"user-agent={CONFIGURATION['USER_AGENT']}")

driver = google_colab_selenium.Chrome(options=options)
```

## Configuración

```python
CONFIGURATION = {
    "BASE_URL": "https://girls4stem.uv.es/#/expertaStem#expertsCards"
}
```

## Request

```python
driver.get(CONFIGURATION["BASE_URL"])
```

## Parseo y extracción

```python
speakers = driver.find_elements(by=By.CLASS_NAME, value="single-speaker-area")
len(speakers)
```

Se podría grabar un vídeo, o incluso tomar capturas de pantalla de ciertas partes del proceso

```python
def get_speaker_info(speaker):
  driver.execute_script("arguments[0].click();", speaker)

  speaker_details = driver.find_element(by=By.CLASS_NAME, value="mat-dialog-container")
  # driver.get_screenshot_as_file("screenshot.png")

  info = {
      "url_imagen_de_perfil": speaker.find_element(by=By.TAG_NAME, value="img").get_attribute("src"),
      "nombre": speaker_details.find_element(by=By.ID, value="name").get_attribute("value").strip(),
      "primer_apellido": speaker_details.find_element(by=By.ID, value="surname").get_attribute("value").strip(),
      "segundo_apellido": speaker_details.find_element(by=By.ID, value="surname2").get_attribute("value").strip(),
      "puesto_de_trabajo": speaker_details.find_element(by=By.ID, value="job").get_attribute("value").strip(),
      "empresa": speaker_details.find_element(by=By.ID, value="company").get_attribute("value").strip(),
      "campo_de_trabajo": speaker_details.find_element(by=By.ID, value="field").get_attribute("value").strip(),
      "biografia": speaker_details.find_element(by=By.ID, value="bibliography").get_attribute("value").strip(),
  }

  # close modal
  backdrop = driver.find_element(by=By.CLASS_NAME, value="cdk-overlay-backdrop")
  driver.execute_script("arguments[0].click();", backdrop)

  return info
```

```python
speakers_info = []

prod_speakers = speakers
total_speakers = len(prod_speakers)

for index, speaker in enumerate(prod_speakers):
  speakers_info.append(get_speaker_info(speaker))
  print(f"({index + 1}/{total_speakers})", "-", round((index + 1) / total_speakers * 100, 2), "%")
```

## Cerramos conexión, por educación

```python
driver.close()
```

## Guardado de datos

Este paso sería el de ingesta de datos

```python
df = pandas.DataFrame(speakers_info)
df
```
